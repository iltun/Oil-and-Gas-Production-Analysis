<b> Overview of the Dataset</b>

Dataset contains detailed records related to oil and gas production. It appears to cover a time span beginning from the year 2001, and includes various parameters associated with oil and gas wells, their production, and operational statuses. This dataset is valuable for analyzing trends in the oil and gas industry, understanding regional production capacities, and assessing the productivity of different wells.

<b>Key Columns and Their Significance</b>

API Well Number: A unique identifier for each well. This number is crucial for distinguishing between different wells in the dataset.

County: Indicates the geographical location of each well at the county level. This information is useful for regional analysis of production and understanding geographical distribution.

Company Name: The name of the company operating the well. It allows analysis of production based on different operators.

Well Type Code: Identifies the type of well. Different types of wells may have varying production capacities and operational characteristics.

Production Field: The field where the well is located, which can be an indicator of geological characteristics.

Well Status Code: Provides the operational status of the well, such as active, inactive, or abandoned.

Production Data (Oil Produced, Gas Produced, Water Produced): These columns contain quantitative measures of production for oil, gas, and water. They are central to any analysis of productivity and efficiency.

Reporting Year: The year for which the production data is reported. This allows for time-series analysis to identify trends over the years.

<b>Insights and Applications</b>

Production Trends: By analyzing the production data over time, one can identify trends, peak production periods, and potential declines in output.

Geographical Analysis: Comparing production across different counties can reveal regional strengths and resource-rich areas.

Company Performance: Analyzing data by company names can provide insights into the performance and contribution of different operators in the industry.

Well Status and Type Analysis: Understanding the distribution of well types and statuses can inform about the operational focus in the industry and potential shifts in extraction methods.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('oil-and-gas-annual-production-beginning-2001-1.csv')
df.head()
```

```python
# Data Cleaning
# data types
df.dtypes
```

```python
# Standardizing column names
df.columns = df.columns.str.replace(' ', '_').str.lower()
```

```python
# Handling Missing Values
# Assessing the extent of missing data
df.isnull().sum()
```

```python
df.head()
```

```python
# Checking for duplicate rows
duplicate_rows = df.duplicated().sum()
duplicate_rows
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# Checking for irrelevant columns
# Columns might be considered irrelevant if they contain a single value (or no variability), or are not useful for analysis
irrelevant_columns = [col for col in df.columns if df[col].nunique() <= 1]
irrelevant_columns
```

```python
# Checking for missing values in each column
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values
```

```python
# Percentage of missing values
missing_percentage = (missing_values / len(df)) * 100
missing_percentage
```

```python
# Selecting numerical columns for outlier detection
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Plotting boxplots for each numerical column to visualize outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns), 1, i + 1)
    sns.boxplot(x=df[col])
    plt.title(col)
```

```python
# Handling missing values
# Imputing missing values for production-related columns with median
for col in ['Oil Produced, bbl', 'Gas Produced, Mcf', 'Water Produced, bbl']:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)
```

```python
# Imputing 'Months in Production' with median
df['Months in Production'].fillna(df['Months in Production'].median(), inplace=True)
```

```python
# Imputing categorical columns with 'Unknown' or mode
for col in ['Town', 'Production Field']:
    if df[col].isnull().sum() > 0:
        df[col].fillna('Unknown', inplace=True)
```

```python
# Filling very small missing values with mode for categorical and median for numerical
small_missing_cols = ['County', 'Well Name', 'Well Status Code', 
                      'Completion Code', 'Sidetrack Code', 'API Hole Number', 
                      'Producing Formation', 'Company Name', 'New Georeferenced Column']
for col in small_missing_cols:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)
```

```python
# Checking if all missing values are handled
missing_values_after = df.isnull().sum().max()
missing_values_after
```

```python
df.isnull().sum()
```

```python
import numpy as np

# Log-transforming production-related columns to reduce the impact of extreme values
# Adding a small constant to avoid log(0)
small_constant = 1e-4
production_cols = ['Oil Produced, bbl', 'Gas Produced, Mcf', 'Water Produced, bbl']
for col in production_cols:
    df[col] = np.log(df[col] + small_constant)

# Handling outliers in other numerical columns using IQR
for col in numerical_columns:
    if col not in production_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        upper_limit = Q3 + 1.5 * IQR
        lower_limit = Q1 - 1.5 * IQR
        df[col] = np.where(df[col] > upper_limit, upper_limit, 
                                     np.where(df[col] < lower_limit, lower_limit, df[col]))

# Verify the changes
df.head()
```

```python
# Focusing on key production columns
production_columns = ['Oil Produced, bbl', 'Gas Produced, Mcf', 'Water Produced, bbl']
# Visualizing the distribution of production data
plt.figure(figsize=(15, 5))

for i, col in enumerate(production_columns):
    plt.subplot(1, len(production_columns), i + 1)
    df[col].hist(bins=50)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Focusing on the "Oil Produced, bbl" column for detailed exploration
oil_produced = df['Oil Produced, bbl']

# Calculating basic summary statistics manually
oil_produced_min = oil_produced.min()
oil_produced_max = oil_produced.max()
oil_produced_mean = oil_produced.mean()
oil_produced_median = oil_produced.median()
oil_produced_std = oil_produced.std()

# Summary statistics
oil_summary = {
    'Minimum': oil_produced_min,
    'Maximum': oil_produced_max,
    'Mean': oil_produced_mean,
    'Median': oil_produced_median,
    'Standard Deviation': oil_produced_std
}

oil_summary
```

```python

# Visualizing the distribution of "Oil Produced, bbl"
plt.figure(figsize=(8, 4))
oil_produced.hist(bins=50)
plt.title('Distribution of Oil Produced (bbl)')
plt.xlabel('Oil Produced (bbl)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Assuming 'data' is your DataFrame and it has columns 'Reporting Year', 'Oil Produced, bbl'
df['Reporting Year'] = pd.to_datetime(df['Reporting Year'], format='%Y')
annual_oil_production = df.groupby(df['Reporting Year'].dt.year)['Oil Produced, bbl'].sum()
plt.figure(figsize=(10, 6))
annual_oil_production.plot(kind='line', marker='o')
plt.title('Annual Oil Production Trends')
plt.xlabel('Year')
plt.ylabel('Total Oil Produced (bbl)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df.boxplot(column='Gas Produced, Mcf')
plt.title('Gas Production Spread')
plt.ylabel('Gas Produced (Mcf)')
plt.show()
```

```python
# Assuming your DataFrame has multiple numerical columns like 'Oil Produced, bbl', 'Gas Produced, Mcf'
correlation_matrix = df[['Oil Produced, bbl', 'Gas Produced, Mcf']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Simulating some data for demonstration
np.random.seed(0)
years = np.arange(2001, 2022)
oil_production = np.random.normal(1000, 200, len(years))  # Simulated oil production data
gas_production = np.random.normal(800, 150, len(years))  # Simulated gas production data
# Time Series Line Graph for annual production trends
plt.figure(figsize=(10, 6))
plt.plot(years, oil_production, label='Oil Production (bbl)', marker='o')
plt.plot(years, gas_production, label='Gas Production (Mcf)', marker='x')
plt.title('Annual Production Trends')
plt.xlabel('Year')
plt.ylabel('Production')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Grouping data by 'County' and calculating average oil production
average_oil_by_county = df.groupby('County')['Oil Produced, bbl'].mean()
# Sorting data for better visualization
average_oil_by_county = average_oil_by_county.sort_values(ascending=False)
plt.figure(figsize=(12, 8))
average_oil_by_county.plot(kind='bar')
plt.title('Average Oil Production by County')
plt.xlabel('County')
plt.ylabel('Average Oil Produced (bbl)')
plt.xticks(rotation=45)
plt.show()
```

```python
well_type_counts = df['Well Type Code'].value_counts()
plt.figure(figsize=(8, 8))
well_type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Well Types')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
# Cumulative production over time
cumulative_production = df.groupby('Reporting Year')[['Oil Produced, bbl', 'Gas Produced, Mcf']].cumsum()
plt.figure(figsize=(12, 8))
plt.plot(cumulative_production.index, cumulative_production['Oil Produced, bbl'], label='Cumulative Oil Production')
plt.plot(cumulative_production.index, cumulative_production['Gas Produced, Mcf'], label='Cumulative Gas Production')
plt.title('Cumulative Production Over Time')
plt.xlabel('Year')
plt.ylabel('Cumulative Production')
plt.legend()
plt.show()
```

```python
# Creating a pivot table for the heatmap
heatmap_data = df.pivot_table(values='Oil Produced, bbl', index='County', columns='Reporting Year', aggfunc='sum')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis')
plt.title('Oil Production Heatmap by County and Year')
plt.xlabel('Year')
plt.ylabel('County')
plt.show()
```

```python
# Reverting the logarithmic transformation
df['Oil Produced, bbl'] = np.exp(df['Oil Produced, bbl'])
# Simple histogram of the reverted data
plt.figure(figsize=(8, 6))
df['Oil Produced, bbl'].hist(bins=30)
plt.title('Distribution of Oil Production (Reverted Scale)')
plt.xlabel('Oil Produced (bbl)')
plt.ylabel('Frequency')
plt.show()
```
